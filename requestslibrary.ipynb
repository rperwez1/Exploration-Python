```python
import requests
```

```python
r = requests.get('https://www.wikihow.com/Sleep-Better')
print(r)
```

```python
#attributes and methods that we can access
print(dir(r))
```

```python
print(help(r))
```

```python
#content of the response in unicode (html)
print(r.text)
```

```python
r = requests.get('https://www.wikihow.com/images/thumb/f/f3/Sleep-Better-Step-1-Version-8.jpg/aid3476-v4-728px-Sleep-Better-Step-1-Version-8.jpg.webp')
with open('sleep.jpeg', 'wb') as f:
    f.write(r.content)
```

```python
# image is now saved in directory where python is
```

```python
#http status code
print(r.status_code)
```

```python
print(r.ok)
```

```python
print(r.headers)
```

```python
payload = {'page': 2, 'count': 25}
bin = requests.get('https://httpbin.org/get', params = payload)
print(bin.text)
```

```python
print(bin.url)
```

```python
#posting form data to route etc...
payload2 = {'username': 'ramsha', 'password': 'hello'}
bin2 = requests.post('https://httpbin.org/post', data = payload2)
print(bin2.text)
```

```python
#json responses
print(bin2.json())
```

```python
bin_dict = bin.json()
print(bin_dict['args'])
```
